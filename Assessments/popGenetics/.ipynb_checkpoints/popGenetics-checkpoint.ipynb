```python
%reset
import numpy as np
import matplotlib.pyplot as plt
from random import randrange
import seaborn as sns
%matplotlib inline
```

```python
def initial_pop(size, ratio):
    allele_1 = int(size*ratio)
    allele_2 = int(size-allele_1)
    A = list('A')*allele_1
    B = list('B')* allele_2
    pop = A+B
    return pop

pop = initial_pop(100,0.5)
pop.count('A')

proportion_A = [pop.count('A')/len(pop)]
proportion_B = [pop.count('B')/len(pop)]
print(proportion_A == proportion_B)

proportion_B[-1]

generation = 1
pop_range = range(len(pop))
```

```python
while generation < 1000:
    pop_next = []
    for i in pop_range:
        random_index = randrange(len(pop))
        pop_next.append(pop[random_index])

    pop = pop_next
    proportion_A.append(pop.count('A')/len(pop))
    proportion_B.append(pop.count('B')/len(pop))
    
    generation += 1
    
    if (proportion_A[-1] == 1 or proportion_A[-1] == 0):
        break
# print(generation)

print(len(proportion_A))
print(generation)
```

```python
# proportion_A
```

```python
sns.set_style("darkgrid")
fig = plt.figure(figsize=(10,6))
ax = plt.axes()
x = np.arange(0,len(range(generation)),1)
line_1 = ax.plot(x, proportion_A)
line_2 = ax.plot(x, proportion_B)
plt.legend('AB')
plt.xlabel('Generations', fontsize=12)
plt.ylabel('Allelic proportion', fontsize=12)
plt.title('Genetic Drift', fontsize=15)
plt.show()
```

```python
pop = []
A_homo = ['A','A']
hetero = ['A','a']
a_homo = ['a','a']


for i in range(25):
    pop.append(A_homo)
    pop.append(a_homo)

for i in range(50):
    pop.append(hetero)

# pop
len(pop)
# pop[-1][0]


homo_A_prop = [pop.count(['A', 'A'])/len(pop)]
print(homo_A_prop)

hetero_prop = [(pop.count(['A', 'a']) + pop.count(['a', 'A']))/len(pop)]
print(hetero_prop)

homo_a_prop = [pop.count(['a', 'a'])/len(pop)]
print(homo_a_prop)
```

```python
generation = 1
while generation <= 501:
    pop_next = []
    while len(pop_next) < 100:
            parent_1_allele = pop[randrange(len(pop))][randrange(2)]
            parent_2_allele = pop[randrange(len(pop))][randrange(2)]
            child = [parent_1_allele] + [parent_2_allele]
            if child[0] == 'a' and child[1] == 'a':
                fitness = randrange(100)
                #print(fitness)
                if fitness < 80:
                    pop_next.append(child)
                else:
                    pass
                    #print('failed: ' + str(child))
            else:
                pop_next.append(child)

    homo_A_prop.append(pop_next.count(['A', 'A'])/len(pop_next))
    hetero_prop.append((pop_next.count(['A', 'a'])+ pop_next.count(['a', 'A']))/len(pop))
    homo_a_prop.append(pop_next.count(['a', 'a'])/len(pop_next))

    pop = pop_next
    generation += 1
    
    if (homo_a_prop[-1] <= 0):
        break

print(generation)
print(len(homo_A_prop))
```
